### Zad 1:

Zaimplementuj funkcję:
$f(x) = \frac{1}{8}x^{4} + 2(x + 0.7)^{3} - \frac{1}{2}x^{2} - 38 x + 1$.



```python
def f(x):
    return (x**4)/8 + 2*(x+0.7)**3 - 1/2*x*x - 38*x + 1

f(0)
```

### Zad 2:

Oblicz wartość funkcji $f(x)$ w przedziale [-20, 20] z krokiem $\Delta = 1$. Wyniki zapisz na liście.



```python
Y = []
for i in range(-20,20):
    Y.append(f(i))
    
Y
```

### Zad 3:

Wyświetl wykres funkcji $f(x)$ korzystając z wyników z zad. 2.

```python
import matplotlib.pyplot as plt

X = range(-20, 20)

plt.figure(figsize=(5,3))
plt.plot(X, Y)
plt.title('Wykres funckji f(x)')
plt.xlabel('Oś x')
plt.ylabel('Oś y')
# plt.xticks([1, 2, 3, 4, 5])
# plt.yticks([1, 2, 3, 4, 5])
# plt.xlim(2.5, 3.5)
# plt.ylim(4, 6.1)
```

### Zad 4:

Korzystając z listy otrzymanej w zad. 2, znajdź takie $x$, dla którego funkcja $f(x)$ przyjmuje najmniejszą wartość. Wyświetl $x$ oraz $f(x)$.

```python
min_value = min(Y)
min_index = Y.index(min(Y))
print('wartość: ' + str(min_value))
print('index: ' + str(min_index))
print('x: ' + str(X[min_index]))
```

### Zad 5:

Wyznacz inny krok ($\Delta' < \Delta$) z zakresu [-20, 20] i sprawdź czy odnajdziesz mniejszą wartość funkcji $f(x)$ niż w poprzednim zadaniu? Jeżeli tak, to podaj $x$ oraz wartość funkcji $f(x)$.

```python
def drange(start, stop, step):
    result = []
    r = start
    while r < stop:
        result.append(r)
        r += step
    return result
        
X_smaller = drange(-20, 20, 0.1)
Y_smaller = []

for i in X_smaller:
    Y_smaller.append(f(i))
    if f(i) < min_value:
        print(str(i) + ' -> ' +  str(f(i)))
        
min(Y_smaller)
```

### Zad 6:

Wyświetl wykres (parametry wykresu dobierz samodzielnie) prezentujący minimum funkcji $f(x)$ znalezione w zad. 5.


```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5,3))
plt.plot(X_smaller, Y_smaller)
plt.title('Minimum funckji f(x)')
plt.xlabel('Oś x')
plt.ylabel('Oś y')
plt.xlim(1.6, 2)
plt.ylim(-36.7, -36)
```

### Zad 7:

Wygeneruj $k$ losowych wartości $x$ w przedziale [-20, 20](metoda Monte Carlo), a następnie znajdź minimum funkcji $f(x)$.
Wyświetl wykres funkcji.

```python
import matplotlib.pyplot as plt
import random

k = 20

x_random = []
for i in range(k): 
    x_random.append(random.uniform(-20, 20))

x_random = sorted(x_random)

y_random = []
for i in range(k):
    y_random.append(f(x_random[i]))
    
print(min(y_random))

plt.figure(figsize=(5,3))
plt.plot(x_random, y_random)
plt.title('Wykres funckji f(x)')
plt.xlabel('Oś x')
plt.ylabel('Oś y')
```
