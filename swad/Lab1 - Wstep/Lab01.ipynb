#### Zadanie 5

Wczytaj dane z pliku `dane-ludnosc.npz`. Plik ten zawiera zbiór danych zawierający **liczbę ludnosci**.

Oblicz wydatki z budżetu na mieszkańca dla każdego z województw (_uwzględnij dane z roku 2008_).


````python
import numpy as np
with open('./dataset/dane-ludnosc.npz', 'rb') as f:
    content = np.load(f)
    columns, data = content['columns'].tolist(), content['data'].tolist()
    
# print(columns)
# data

people_in_2008 = [ [row[1]] + [float(row[5])] for row in data if row[4] == '2008' and row[1] != 'POLSKA']
money_in_2008 = [ [row[1]] + [float(row[4])] for row in data_from_2008 if row[3] == '2008' and row[1] != 'POLSKA']

zipped = []

for woj1, wartosc1 in people_in_2008:
    for woj2, wartosc2 in money_in_2008:
        if woj1 == woj2:
            zipped.append([woj1, wartosc1, wartosc2])

# print(zipped)
for w in zipped:
    print(w[0] + ' ' + str(round(w[2] / w[1], 2)))
````

#### Zadanie 6

Które województwo odnotowało największy (oraz najmniejszy) wzrost wydatków z budżetu na mieszkańca pomiędzy rokiem 2002 a rokiem 2020?

Oblicz współczynnik wydatków z budżetu na mieszkańca jako stosunek wartości z roku 2020 do wartości z roku 2002.

$$
wspolczynnik = \frac{w_{2020}}{w_{2002}}
$$

gdzie $w_{2020}$ to wydatki z budżetu na mieszkańca w roku 2020, a $w_{2002}$ to wydatki z budżetu na mieszkańca w roku 2002.


````python
import numpy as np
with open('./dataset/dane-ludnosc.npz', 'rb') as f:
    content = np.load(f)
    _, data_people = content['columns'].tolist(), content['data'].tolist()
    
with open('./dataset/dane.npz', 'rb') as f:
    content = np.load(f)
    _, data_money = content['columns'].tolist(), content['data'].tolist()

data_money = [[row[1]] + [row[3]] +  [float(row[4])] for row in data_money if (row[3] == '2002' or row[3] == '2020') and row[1] != 'POLSKA']
data_people = [[row[1]] + [row[4]] + [float(row[5])] for row in data_people if (row[4] == '2002' or row[4] == '2020') and row[1] != 'POLSKA']

connected = []
for state1, year1, v1 in data_money: 
    for state2, year2, v2 in data_people: 
        if state1 == state2 and year1 == year2:
            connected.append([state1, year1, v1/v2])
            
result = []

for state1, year1, v1 in connected:
    if year1 == '2002':  
        for state2, year2, v2 in connected:
            if state1 == state2 and year2 == '2020': 
                result.append([state1, v2 / v1])

result = sorted(result, key=lambda row: row[1], reverse=True)

print(result[0])
print(result[-1])
````
